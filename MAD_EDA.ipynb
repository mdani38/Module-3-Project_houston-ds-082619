```python
import seaborn as sns
import pandas as pd
import sqlite3
conn = sqlite3.connect('database.sqlite')
cur = conn.cursor()
```

```python
cur.execute("""SELECT *
                FROM reviews R
                JOIN genres G 
                USING(reviewid)
                JOIN labels L 
                USING (reviewid)
                JOIN content c
                USING (reviewid)
               ;""")



df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```
